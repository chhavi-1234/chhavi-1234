```python
import matplotlib.pyplot as plt
import pandas as pd
import requests
from PIL import Image
from io import BytesIO

# Assuming df is already loaded
# Example: df = pd.read_csv("movie_genres_screenshots_120.csv")

# Number of images to show
N = 20  # You can increase this to 50, 100, etc.

for i in range(N):
    try:
        url = df.loc[i, 'ImageURL']
        title = df.loc[i, 'MovieName']
        genres = df.loc[i, 'Genres'].replace('|', ', ')
        
        # Fetch image
        response = requests.get(url, timeout=5)
        img = Image.open(BytesIO(response.content))

        # Plot image
        plt.imshow(img)
        plt.axis('off')
        plt.title(f"{title}\n({genres})", fontsize=10)
        plt.show()

    except Exception as e:
        print(f"Error displaying image {i}: {e}")
```
